```python
import os
import shutil
import tensorflow as tf
from tensorflow.keras import models, layers
from sklearn.utils import resample
import matplotlib.pyplot as plt

# Constants
IMAGE_SIZE = 224
CHANNELS = 3
BATCH_SIZE = 32
EPOCHS = 10

# Directory paths
dataset_dir = 'training'  # Ensure this path points to the correct data folder
output_dir = 'Balanced_Data__is'  # New folder for balanced datasets
images_dir = "Images"  # Folder containing test images

# Function to balance dataset by upsampling
def balance_dataset(images, target_count):
    if not images:
        return []  # Return an empty list if there are no images
    if len(images) >= target_count:
        return resample(images, n_samples=target_count, replace=False, random_state=42)
    else:
        return resample(images, n_samples=target_count, replace=True, random_state=42)

# Function to split images by class and balance the dataset
def prepare_balanced_dataset_per_class(dataset_dir, output_dir, target_count=500):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    for class_name in os.listdir(dataset_dir):
        class_dir = os.path.join(dataset_dir, class_name)
        if not os.path.isdir(class_dir):
            continue

        # Collect images for the current class and for "not this class"
        class_images = [os.path.join(class_dir, img) for img in os.listdir(class_dir) if img.lower().endswith(('.png', '.jpg', '.jpeg'))]
        if not class_images:
            print(f"Skipping class '{class_name}' as it has no images.")
            continue

        non_class_images = [os.path.join(root, img) for root, _, files in os.walk(dataset_dir) for img in files if root != class_dir and img.lower().endswith(('.png', '.jpg', '.jpeg'))]

        # Balance both positive and negative images for the current class
        balanced_class_images = balance_dataset(class_images, target_count)
        balanced_non_class_images = balance_dataset(non_class_images, target_count)

        # Skip if no balanced data could be created
        if not balanced_class_images or not balanced_non_class_images:
            print(f"Skipping class '{class_name}' due to insufficient images.")
            continue

        # Create directories to store balanced dataset for this class
        class_output_dir = os.path.join(output_dir, class_name)
        if not os.path.exists(class_output_dir):
            os.makedirs(class_output_dir)

        positive_dir = os.path.join(class_output_dir, 'positive')
        negative_dir = os.path.join(class_output_dir, 'negative')
        os.makedirs(positive_dir, exist_ok=True)
        os.makedirs(negative_dir, exist_ok=True)

        # Copy balanced images to appropriate folders
        for img in balanced_class_images:
            shutil.copy(img, positive_dir)
        for img in balanced_non_class_images:
            shutil.copy(img, negative_dir)

        print(f"Balanced dataset for class '{class_name}' with {target_count} positive and negative images each.")

# Prepare balanced datasets
prepare_balanced_dataset_per_class(dataset_dir, output_dir)

# Define a function to create and train a model for each class
def train_model_for_class(class_name):
    class_path = os.path.join(output_dir, class_name)
    if not os.path.exists(class_path) or not os.path.isdir(class_path):
        print(f"Directory for class '{class_name}' not found. Skipping training.")
        return None

    # Load the balanced dataset for this class
    try:
        train_dataset = tf.keras.preprocessing.image_dataset_from_directory(
            class_path,
            label_mode='binary',  # Binary classification for this class
            batch_size=BATCH_SIZE,
            image_size=(IMAGE_SIZE, IMAGE_SIZE),
            validation_split=0.2,
            subset="training",
            seed=42
        )
        val_dataset = tf.keras.preprocessing.image_dataset_from_directory(
            class_path,
            label_mode='binary',  # Binary classification for this class
            batch_size=BATCH_SIZE,
            image_size=(IMAGE_SIZE, IMAGE_SIZE),
            validation_split=0.2,
            subset="validation",
            seed=42
        )
    except Exception as e:
        print(f"Error loading dataset for class '{class_name}': {e}")
        return None

    # Prefetch for performance
    train_dataset = train_dataset.prefetch(buffer_size=tf.data.AUTOTUNE)
    val_dataset = val_dataset.prefetch(buffer_size=tf.data.AUTOTUNE)

    # Define the model
    model = models.Sequential([
        layers.Input(shape=(IMAGE_SIZE, IMAGE_SIZE, CHANNELS)),
        layers.Rescaling(1./255),
        layers.RandomFlip("horizontal_and_vertical"),
        layers.RandomRotation(0.2),
        layers.Conv2D(32, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(64, activation='relu'),
        layers.Dense(1, activation='sigmoid'),  # Binary classification
    ])

    # Compile the model
    model.compile(
        optimizer='adam',
        loss=tf.keras.losses.BinaryCrossentropy(),
        metrics=['accuracy']
    )

    # Train the model
    history = model.fit(
        train_dataset,
        epochs=EPOCHS,
        validation_data=val_dataset,
        callbacks=[
            tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6),
            tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
        ]
    )

    # Save the trained model
    model.save(f"{class_name}_model.keras")

    # Plot training history
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'Training Accuracy for class {class_name}')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()

    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'Training Loss for class {class_name}')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

    return model

# Train a separate model for each class
class_models = {}
for class_name in os.listdir(dataset_dir):
    if os.path.isdir(os.path.join(dataset_dir, class_name)):
        print(f"\nTraining model for class: {class_name}")
        model = train_model_for_class(class_name)
        if model:
            class_models[class_name] = model

# Function to predict and classify images in 'images' folder
def classify_images(models, images_dir):
    for img_name in os.listdir(images_dir):
        img_path = os.path.join(images_dir, img_name)
        if not img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
            continue

        # Preprocess image
        img = tf.keras.preprocessing.image.load_img(img_path, target_size=(IMAGE_SIZE, IMAGE_SIZE))
        img_array = tf.keras.preprocessing.image.img_to_array(img)
        img_array = tf.expand_dims(img_array, 0)  # Create a batch

        # Predict with each model and find the highest accuracy
        predictions = {}
        for class_name, model in models.items():
            pred = model.predict(img_array)[0][0]
            predictions[class_name] = pred

        # Find predicted class based on highest prediction confidence
        predicted_class = max(predictions, key=predictions.get)
        print(f"Image '{img_name}' predicted as '{predicted_class}' with confidence {predictions[predicted_class]:.2f}")

        # Copy image to the predicted class folder
        target_dir = os.path.join(output_dir, predicted_class)
        os.makedirs(target_dir, exist_ok=True)
        shutil.copy(img_path, os.path.join(target_dir, img_name))

# Classify images in 'images' folder
classify_images(class_models, images_dir)
```
